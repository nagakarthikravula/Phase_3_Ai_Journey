```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "The dog chased the ball",
    "A puppy ran after a sphere",
    "I love eating pizza",
    "She enjoys having pasta for dinner",
    "Machine learning is a subset of AI",
    "I am listening to music by edhsheeran",
    "I am playing ed sheeran song on spotify",
    "I love listening to music"
]

embeddings = model.encode(sentences)

print("Shape of one embedding:", embeddings[0].shape)
print("First 5 numbers of sentence 1 embedding:", embeddings[0][:5])

pairs = [
    (0,1),
    (2,3),
    (0,2),
    (0,4),
    (2,4),
    (5,6),
    (2,6),
    (2,7)
]

for i,j in pairs:
  score = util.cos_sim(embeddings[i],embeddings[j])
  print(f"Sentence {i+1} vs Sentence {j+1}: {score.item(): .4f}")
  print(f"  '{sentences[i]}")
  print(f"  '{sentences[j]}")
  print()
```
